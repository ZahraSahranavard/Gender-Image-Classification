```python
model= keras.models.load_model('fine_tuningl.keras')
```

```python
images = ["Harry_Potter_character_poster.jpg", "images.jpg", "LunaLovegood.jpg"]
all_images = []
for img in images:
    load = image.load_img(img, target_size=(180, 180))
    x = image.img_to_array(load)
    x = np.expand_dims(x, axis=0)
    all_images.append(x)
```

```python
plt.imshow(np.squeeze(all_images[0]).astype('uint8'))
```

```python
plt.imshow(np.squeeze(all_images[1]).astype('uint8'))
```

```python
plt.imshow(np.squeeze(all_images[2]).astype('uint8'))
```

```python
pred1 = model.predict(all_images[0])
pred2 = model.predict(all_images[1])
pred3 = model.predict(all_images[2])
pred4 = model.predict(all_images[3])
```

```python
print(pred1)
print(pred2)
print(pred3)
print(pred4)
```

```python
plt.imshow(np.squeeze(all_images[0]).astype('uint8'))
plt.axis("off")
plt.title("Man" if pred1<0.5 else "Woman")
plt.show()
```
